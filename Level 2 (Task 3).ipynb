# Plotting the locations of restaurants on a map using longitude and latitude coordinates

```python
restaurant_names = df["Restaurant Name"]
latitude = df["Latitude"]
longitude = df["Longitude"]
```

```python
X = df[["Latitude", "Longitude"]]
num_clusters = 5
```

```python
kmeans = KMeans(n_clusters = num_clusters, n_init=10, random_state=40)
df["cluster"] = kmeans.fit_predict(X)
```

```python
map_center = [latitude.mean(), longitude.mean()]
```

```python
restaurant_map = folium.Map(location = map_center, zoom_starts = 12)
```

```python
cluster_color = ["red", "black", "blue", "green", "purple"]
```

```python
for index, row in df.iterrows():
    restaurant_name = row["Restaurant Name"]
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    cuisines = row["Cuisines"]
    rating = row["Aggregate rating"]
    cluster = row["cluster"]
```

```python
popup_text = f"Restaurant: {restaurant_name}\nCuisines: {cuisines}|nRating: {rating}"
```

```python
marker = folium.Marker([latitude, longitude], popup = popup_text)
```

```python
marker.add_to(restaurant_map)
```
