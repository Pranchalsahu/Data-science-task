# Identify Restaurants chain Present In Dataset

```python
duplicate_names=df[df.duplicated(subset=['Restaurant Name'],keep=False)]
```

```python
chain_restaurants=df[df['Restaurant Name'].isin(duplicate_names['Restaurant Name'])]
```

```python
chain_names=chain_restaurants['Restaurant Name'].unique()
print("Restaurant Chains:")
for chain_name in chain_names:
    print(chain_name)
```

# The Rating and Popularity of different restaurant chains

```python
chain_stats=df.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum' }).reset_index()
```

```python
sorted_chains_by_rating=chain_stats.sort_values(by='Aggregate rating', ascending=False) 
```

```python
sorted_chain_by_votes=chain_stats.sort_values(by='Votes',ascending=False)
```

```python
print("Top_Rated Chain:")
print(sorted_chains_by_rating.head(10))
```

```python
print("Most Popular Chains by Total Votes:")
print(sorted_chain_by_votes.head(10))
```
