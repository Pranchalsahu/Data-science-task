# Price Range Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/sahuc/OneDrive/Desktop/Dataset.csv')
```

```python
df
```

# Creating Bar Chart To Visualize The Distribution Of Price Among The Restaurant

```python
plt.figure(figsize=(5.4,6.4))
df['Price range'].value_counts().sort_index().plot(kind='bar',color='black')
plt.xlabel('Price range')
plt.ylabel('Number of Restaurants')
plt.title('Distribution Of Prize Range Among The Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y',linestyle='',alpha=0.6)
plt.show()
```

# Calculate The Percentage Of Restaurants In Each Price Range Category

```python
total_restaurants = len(df)
```

```python
price_range_counts = df['Price range'].value_counts()
```

```python
percentage_restaurants = (price_range_counts / total_restaurants) * 100
```

```python
print("The Percentage Of Restaurants In Each Price Range Category:")
print(percentage_restaurants)
```
