# Price Range Vs Online Delivery And Table Booking

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/sahuc/OneDrive/Desktop/Dataset.csv")
```

```python
df.head()
```

# Analyze Relationship Between Price Range And Online Delivery 

```python
sns.catplot(x="Price range", hue="Has Online delivery" ,kind="count", data=df)
plt.xlabel('Price range')
plt.ylabel('Number Of Restaurants')
plt.title('Relationship Between Price Range And Online Delivery')
plt.show()
```

# Analyze Relationship Betwwen Price Range And Table Booking

```python
sns.catplot(x="Price range", hue="Has Table booking" ,kind="count", data=df)
plt.xlabel('Price range')
plt.ylabel('Number Of Restaurants')
plt.title('Relationship Between Price Range And Table Booking')
plt.show()
```
